Problema 01: Hacer un problema que imprima los números del -100 al 200.

Para que pueda imprimirme todos los números que van desde el -100 al 200, use for con la función range(-100,201), con el fin de que genere todos los números automáticamente. Agregué el 201, ya que la serie solo mostraría hasta el 199.

````python
for number in range(-100, 201):
  print(number)
````

Problema 02: Hacer un programa que imprima los números del -100 al 200 que sean divisibles entre 6 (con resto 0).

Primero pensé en el rango de números que me pedía el problema, desde -100 hasta 200, y use range(-100,201) para recorrerlos con un ciclo for. Después a cada número le aplique el operador módulo % para verificar si era divisible por 6 (es decir, que el resto fuera 0). Finalmente, con el condicional if hice que solo se imprimieran aquellos números que cumplían esa condición.



````python
for number in range(-100, 201):
  if number % 6 == 0:
    print(number)
````

Problema 03: Pseudocódigo, Sumador y comparador de dos números Realiza el algoritmo para un sumador de dos números. Si el resultado es menor a 200 se mostrará el mensaje "menor a 200", si el resultado es mayor a 200, pero menor a 250 mostrará el mensaje "mayor a 200", pero si es mayor a 250 mostrará el mensaje "mayor a 250".

Primero pedí al usuario que ingresara dos números y los guardé en variables. Luego sumé ambos valores y guardé el resultado. Con un condicional if- elif-else comparé esa suma: si era menor que 200 mostré un mensaje, si estaba entre 200 y 250 mostré otro, y si era mayor a 250 mostré el mensaje correspondiente. Finalmente, imprimí el resultado.

````python
numero1 = int(input("Ingrese el primer numero "))
numero2 = int(input("Ingrese el segundo numero"))

suma = numero1 + numero2
resultado = ""
if suma < 200:
  resultado = "menor a 200"
elif suma > 200 and suma < 250:
  resultado = "mayor a 200"
else:
  resultado = "mayor a 250"

print(resultado)

````

Problema 04: Generar un programa que combine variables, tipos de datos y una condicional en Python (hacer una función que evalúe si el usuario es mayor de edad y si le gusta la música urbana, que envíe diferentes textos según la respuesta del usuario)

Primero utilicé input() para pedir la edad y la convertí a número con int() para poder compararla; luego pedí si le gusta la música urbana y apliqué .lower () para estandarizar la respuesta. Después use condicionales if y elif junto con operadores lógicos para evaluar las dos variables al mismo tiempo, contemplando los cuatro casos posibles (mayor/menor edad y con/sin gusto musical). Según el resultado, asigné un mensaje a la variable resultado y finalmente lo mostré en pantalla con print().
```
# Esto tiene formato de código
```



````python
edad = int(input("cual es tu edad: "))
gusto_musica_urbana = input("te gusta la musica urbana 'si' o 'no': ").lower()

if edad >= 18 and gusto_musica_urbana == "si":
    resultado = "es mayor de edad y le gusta la musica urbana"
elif edad >= 18 and gusto_musica_urbana == "no":
    resultado = "es mayor de edad y no le gusta la musica urbana"
elif edad < 18 and gusto_musica_urbana == "si":
    resultado = "es menor de edad y le gusta la musica urbana"
elif edad < 18 and gusto_musica_urbana == "no":
    resultado = "es menor de edad y no le gusta la musica urbana"

print(resultado)


````
